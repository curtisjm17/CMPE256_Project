## Import Required Libraries 

```python
import json
import numpy as np
import time
import os
import pandas as pd
import csv
```

## Read in Spotify Data 

```python
# List of all input data files
files = []
for file in os.listdir(".\data\spotify_million_playlist_dataset\data"):
    if file.endswith(".json"):
        files.append(file)
```

```python
# Initialize variables to store artists, tracks and albums
artist_all = []
track_all = []
album_all = []

# Loop through each file and store data
count = 1
t0 = time.time()
for file in files:
    
    # Open file
    f = open("./data/spotify_million_playlist_dataset/data/" + file)

    # Read JSON data
    data = json.load(f)

    # Closing file
    f.close()
    
    # Loop through each playlist in file
    for playlist in data['playlists']:

        # Loop through each track in playlist
        for song in playlist['tracks']:
            
            # Split text strings
            artist = song['artist_uri'].split(':')
            track = song['track_uri'].split(':')
            album = song['album_uri'].split(':')
    
            # Store all artists, tracks and albums
            artist_all.append(artist[2])
            track_all.append(track[2])
            album_all.append(album[2])
            
    # Eliminate all duplicate entries
    artist_all = list(set(artist_all))
    album_all = list(set(album_all))
    track_all = list(set(track_all))
    
    # Display progress
    print('Filename:', file)
    print('File ', count, ": ", count/len(files), 'Time: ', time.time()-t0)
    count = count + 1
```

## Produce CSV Files of lists of Artists, Tracks and Albums 

```python
# Initialize list to store artists, albums and tracks as list of list
artist = []
album = []
track = []

# Converts artist list to list of lists
for ele in artist_all:
    artist.append([ele])

# Converts album list to list of lists
for ele in album_all:
    album.append([ele])

# Converts track list to list of lists
for ele in track_all:
    track.append([ele])
```

```python
# Write artists to csv file
with open('./data/artists.csv', 'w', newline='') as f: 
    write = csv.writer(f) 
    write.writerows(artist) 

# Write albums to csv file
with open('./data/albums.csv', 'w', newline='') as f: 
    write = csv.writer(f) 
    write.writerows(album) 

# Write tracks to csv file
with open('./data/tracks.csv', 'w', newline='') as f: 
    write = csv.writer(f) 
    write.writerows(track) 
```
