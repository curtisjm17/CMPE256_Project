## Import Required Libraries 

```python
import json
import numpy as np
import time
import os
import pandas as pd
import csv
import random
from itertools import combinations
```

## Read in Spotify Data 

```python
# List of all input data files
files = []
for file in os.listdir(".\data\spotify_million_playlist_dataset\data"):
    if file.endswith(".json"):
        files.append(file)
```

```python
# Initialize lists to store artist, playlists and tracks
artist_playlists = dict({})
track_playlists = dict({})
album_playlists = dict({})

# Initialize variables to store artists, tracks and albums
artist_2_id = dict([])
track_2_id = dict([])
album_2_id = dict([])
track_name_all = dict([])

# Initialize artist, track and album counts
artist_count = 0
track_count = 0
album_count = 0
playlist_count = 0

# Loop through each file and store data
count = 1
t0 = time.time()
for file in files:
    
    # Open file
    f = open("./data/spotify_million_playlist_dataset/data/" + file)

    # Read JSON data
    data = json.load(f)

    # Closing file
    f.close()
    
    # Loop through each playlist in file
    for playlist in data['playlists']:
        
        # Initialize list of artists, traacks and albums for playlist
        artist_list = []
        track_list = []
        album_list = []

        # Loop through each track in playlist
        for song in playlist['tracks']:
            
            # Split text strings for artist, track and album
            artist = song['artist_uri'].split(':')
            track = song['track_uri'].split(':')
            album = song['album_uri'].split(':')
            
            # Get name of artist, track and album
            track_name = song['track_name']
            artist_name = song['artist_name']
            album_name = song['album_name']
    
            # Store all artists
            artist_id = artist_2_id.get(artist[2], -1)
            if artist_id < 0:
                artist_2_id[artist[2]] = artist_count
                artist_count = artist_count + 1
            artist_list.append(artist_2_id[artist[2]])
            
            # Store all track
            track_id = track_2_id.get(track[2], -1)
            if track_id < 0:
                track_2_id[track[2]] = track_count
                track_count = track_count + 1
            track_list.append(track_2_id[track[2]])
            
            # Store all albums
            album_id = album_2_id.get(album[2], -1)
            if album_id < 0:
                album_2_id[album[2]] = album_count
                album_count = album_count + 1
            album_list.append(album_2_id[album[2]])
            
            # Store artist and track name
            track_name_all[track[2]] = [track_name + ' - ' + artist_name]
            
        # Append playlist on list of all playlists
        artist_playlists[playlist_count] = np.unique(artist_list)
        track_playlists[playlist_count] = np.unique(track_list)
        album_playlists[playlist_count] = np.unique(album_list)
        
        # Increase playlist number
        playlist_count = playlist_count+1                     
    
    # Display progress
    print('Filename:', file)
    print('File ', count, ": ", count/len(files), 'Time: ', time.time()-t0)
    count = count + 1
```

## Produce CSV Files of lists of Artists, Tracks and Albums 

```python
# Initialize list to store artists, albums and tracks as list of list
artist = []
track = []
album = []

track_w_name = []

# Converts artist dictionary to list of lists
for key, value in artist_2_id.items():
    artist.append([value, key])

# Converts track dictionary to list of lists
track = []
for key, value in track_2_id.items():
    track.append([value, key])
    
# Converts album dictionary to list of lists
album = []
for key, value in album_2_id.items():
    album.append([value, key])

# Converts track name dictionary to list of lists
track_w_name = []
for key, value in track_name_all.items():
    track_id = track_2_id[key]
    track_w_name.append([key, value[0]])
```

```python
# Write artists to csv file
with open('./data/artists.csv', 'w', newline='') as f: 
    write = csv.writer(f) 
    write.writerows(artist) 

# Write albums to csv file
with open('./data/albums.csv', 'w', newline='') as f: 
    write = csv.writer(f) 
    write.writerows(album) 

# Write tracks to csv file
with open('./data/tracks.csv', 'w', newline='') as f: 
    write = csv.writer(f) 
    write.writerows(track) 
    
# Write tracks with names to csv file
with open('./data/tracks_w_names.csv', 'w', newline='', encoding="utf-8") as f: 
    write = csv.writer(f) 
    write.writerows(track_w_name) 
```

```python
# Write Playlist-Artist ID matrix to csv
with open('./data/artist_w_playlist_id.csv', 'w', newline='') as f:
    c = csv.writer(f)
    count = 0
    c.writerow(['Playlist ID', 'Artist ID'])
    for key in artist_playlists:
        for value in artist_playlists[key]:
            c.writerow([key, value])
            count = count+1
            
# Write Playlist-Track ID matrix to csv
with open('./data/track_w_playlist_id.csv', 'w', newline='') as f:
    c = csv.writer(f)
    count = 0
    c.writerow(['Playlist ID', 'Track ID'])
    for key in track_playlists:
        for value in track_playlists[key]:
            c.writerow([key, value])
            count = count+1

# Write Playlist-Album ID matrix to csv
with open('./data/album_w_playlist_id.csv', 'w', newline='') as f:
    c = csv.writer(f)
    count = 0
    c.writerow(['Playlist ID', 'Album ID'])
    for key in album_playlists:
        for value in album_playlists[key]:
            c.writerow([key, value])
            count = count+1
```
